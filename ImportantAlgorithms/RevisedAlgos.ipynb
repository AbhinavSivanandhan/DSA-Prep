```python
from collections import Counter, deque
import heapq
from typing import List

class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        # Count the frequency of each task
        count = Counter(tasks)
        print(f"Task frequencies: {count}")

        # Create a max-heap based on task frequencies (negative values because Python's heapq is a min-heap)
        maxHeap = [-cnt for cnt in count.values()]
        heapq.heapify(maxHeap)  # Transform the list into a heap
        print(f"Initial maxHeap: {maxHeap}")

        time = 0  # Keeps track of the total time
        q = deque()  # Queue to store tasks during their cooldown: pairs of [remaining_count, next_available_time]

        # Continue while there are tasks in the heap or tasks waiting in cooldown
        while maxHeap or q:
            time += 1  # Increment the time by 1 for each unit of processing
            print(f"\nTime: {time}")

            # If there are no tasks in the maxHeap, fast-forward time to the next available task in the queue
            if not maxHeap:
                print(f"MaxHeap is empty, fast-forwarding time to {q[0][1]}")
                time = q[0][1]
            else:
                # Pop the most frequent task from the heap (invert the negative count back to positive)
                cnt = 1 + heapq.heappop(maxHeap)  # Decrease its count by 1 (since we're processing it)
                print(f"Processing task with remaining count: {-cnt + 1}")
                if cnt:  # If the task still has remaining instances
                    # Append it to the queue with the next time it can be scheduled
                    q.append([cnt, time + n])
                    print(f"Task added to cooldown queue with remaining count {cnt} and available time {time + n}")

            # Check if the task at the front of the queue has finished its cooldown
            if q and q[0][1] == time:  # If the next available time matches the current time
                ready_task = q.popleft()
                heapq.heappush(maxHeap, ready_task[0])  # Push it back to the heap
                print(f"Task with remaining count {ready_task[0]} is back in the maxHeap")

            # Debugging state of maxHeap and queue
            print(f"Current maxHeap: {maxHeap}")
            print(f"Cooldown queue: {list(q)}")

        print(f"\nTotal time required: {time}")
        return time  # Return the total time taken
```

```python
from collections import Counter, deque
import heapq
from typing import List

class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        # Count the frequency of each task
        count = Counter(tasks)
        print(f"Task frequencies: {count}")

        # Create a max-heap with negative values (maxHeap simulates a priority queue)
        maxHeap = [-cnt for cnt in count.values()]
        heapq.heapify(maxHeap)
        print(f"Initial maxHeap (negative counts for max-heap simulation): {maxHeap}")

        time = 0  # Keeps track of the total time
        q = deque()  # Queue to store tasks during their cooldown period: [remaining_count, next_available_time]

        # Continue while there are tasks to process (in heap or queue)
        while maxHeap or q:
            time += 1  # Time progresses by 1 unit
            print(f"\nTime {time} begins...")

            # If there are tasks in the maxHeap, process the most frequent one
            if maxHeap:
                cnt = 1 + heapq.heappop(maxHeap)  # Decrease task count by 1
                print(f"Processing a task, remaining count: {-cnt + 1}")
                if cnt:  # If there are still instances of this task
                    q.append([cnt, time + n])  # Add it to the cooldown queue
                    print(f"Task added to cooldown queue. Cooldown until time {time + n}.")
            else:
                print("No tasks available to process. CPU is idle.")

            # Check if a task's cooldown has finished
            if q and q[0][1] == time:  # If the front task in the queue is ready
                task_ready = q.popleft()
                heapq.heappush(maxHeap, task_ready[0])  # Push it back to the heap
                print(f"Task with count {task_ready[0]} is back in the maxHeap.")

            # Print the state of maxHeap and the cooldown queue at the end of the time unit
            print(f"Current maxHeap: {maxHeap}")
            print(f"Cooldown queue: {list(q)}")

        print(f"\nTotal time required: {time}")
        return time
```

```python
tasks = ["A", "A", "A", "B", "B", "B"]
n = 2
solution = Solution()
solution.leastInterval(tasks, n)
```
