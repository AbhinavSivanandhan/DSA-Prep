```python
def permutations(nums):
    def dfs(path, used, depth):
        # Debug: Entering the function
        print(f"{'  ' * depth}Entering dfs(path={path}, used={used}, depth={depth})")
        
        if len(path) == len(nums):  # Base case: If the path contains all numbers
            result.append(path[:])  # Add a copy of the path to results
            print(f"{'  ' * depth}Complete permutation found: {path}")
            print(f"{'  ' * depth}Returning from dfs")
            return

        # Loop over every number in nums
        for i in range(len(nums)):
            print(f"{'  ' * depth}Checking nums[{i}] = {nums[i]}")

            if used[i]:  # Skip this number if it's already used
                print(f"{'  ' * depth}nums[{i}] = {nums[i]} is already used, skipping...")
                continue
            
            # Choose the current number
            print(f"{'  ' * depth}Choosing nums[{i}] = {nums[i]}")
            used[i] = True
            path.append(nums[i])
            print(f"{'  ' * depth}Path after choosing: {path}, Used array: {used}")
            
            # Recurse with the updated path and used array
            print(f"{'  ' * depth}Recursing into dfs...")
            dfs(path, used, depth + 1)
            
            # Backtrack: Undo the last choice
            print(f"{'  ' * depth}Backtracking: Removing nums[{i}] = {nums[i]} from path")
            path.pop()
            used[i] = False
            print(f"{'  ' * depth}Path after backtracking: {path}, Used array: {used}")
        
        # Debug: Exiting the function
        print(f"{'  ' * depth}Exiting dfs(path={path}, used={used}, depth={depth})")

    result = []
    print(f"Start DFS on nums={nums}")
    dfs([], [False] * len(nums), 0)
    print(f"All permutations: {result}")
    return result

# Example Usage
nums = [1, 2, 3]
print(permutations(nums))
```
