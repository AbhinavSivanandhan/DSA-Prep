```python
def permutations(nums):
    def dfs(path, used, depth):
        # Debug: Show current path and used array
        print(f"{'  ' * depth}dfs(path={path}, used={used})")
        
        if len(path) == len(nums):  # Base case: complete permutation
            result.append(path[:])  # Add a copy of the path to results
            print(f"{'  ' * depth}Complete permutation: {path}")
            return

        for i in range(len(nums)):
            if used[i]:  # Skip already used elements
                print(f"{'  ' * depth}Skip nums[{i}] = {nums[i]} (already used)")
                continue
            
            # Choose the current number
            used[i] = True
            path.append(nums[i])
            print(f"{'  ' * depth}Choose nums[{i}] = {nums[i]}, path: {path}, used: {used}")
            
            # Recurse deeper
            dfs(path, used, depth + 1)
            
            # Backtrack: undo the last choice
            path.pop()
            used[i] = False
            print(f"{'  ' * depth}Backtrack nums[{i}] = {nums[i]}, path: {path}, used: {used}")

    result = []
    print(f"Start DFS on nums={nums}")
    dfs([], [False] * len(nums), 0)
    print(f"All permutations: {result}")
    return result

# Example Usage
nums = [1, 2, 3]
print(permutations(nums))
```
